```python
# Import required libraries 

# Data Manipulation 
import pandas as pd 
import numpy as np 

# Visualization 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
# Load dataset 
store = pd.read_csv("global superstore.csv")
```

```python
# Display dataset shape 
print("Dataset Shape:", store.shape)
```

## DATA CLEANING AND PREPROCESSING

```python
# Convert Date Columns
store['Order.Date'] = pd.to_datetime(store['Order.Date'])

store['Ship.Date'] = pd.to_datetime(store['Ship.Date'])
```

```python
# Check missing values 
store.isnull().sum()
```

```python
# Check duplicates 
store.duplicated().sum()
```

## EXPLORATORY DATA ANALYSIS (EDA) 

```python
# Sales Distribution 
sns.histplot(store['Sales'],
             bins=30,
             kde=True)
plt.show()
```

```python
# Profit Distribution 
sns.histplot(store['Profit'],
             bins=30,
             kde=True)
plt.show()
```

```python
# Sales by Region 
region_sales = store.groupby('Region')['Sales'].sum()

region_sales.plot(kind='bar')

plt.show()
```

```python
# Top Customers 
top_customers = (store.groupby('Customer.Name')['Sales']
    .sum()
    .sort_values(ascending=False)
    .head(10))

top_customers.plot(kind='bar')

plt.show()
```

## 4. Customer Analysis

## 5. Product and Category Insights
